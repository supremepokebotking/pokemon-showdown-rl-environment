```python
results = {'player':0, 'agent':0}
eps_steps = []
eps_rewards = []
for i in range(1):
    team1 = get_random_pokemon_team()
    team2 = get_random_pokemon_team()
    player = Player(team1)
    random_agent = RandomAgent(team2)
    state = State(player, random_agent)
    winner, steps, rewards = state.simulate_battle()
    results[winner] += 1
    eps_steps.append(steps)
    eps_rewards.append(rewards)
```
